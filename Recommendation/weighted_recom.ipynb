```python
import pandas as pd
import numpy as np
```

```python
credits_df = pd.read_csv('C:/Users/shash/Downloads/tmdb-movie-metadata/tmdb_5000_credits.csv')
movies_df = pd.read_csv('C:/Users/shash/Downloads/tmdb-movie-metadata/tmdb_5000_movies.csv')
```

```python
credits_df.head()
movies_df.head()
```

```python
#rename movie_id column in movies_df to id
credits_column_rename = credits_df.rename(index= str, columns = {"movie_id": "id"})
credits_column_rename.head()
```

```python
#merging two data frames based on id

movies_df_merged = movies_df.merge(credits_column_rename, on="id")
movies_df_merged.columns
movies_df_merged.head()
```

```python
#cleaning data
#remove the fields which are not required 
features = ['homepage', 'title_x', 'title_y', 'production_companies']

movies_cleaned_df = movies_df_merged.drop(columns=features)

movies_cleaned_df.columns
```

```python
#using weighted average technique for recommendation (features used : vote_count and vote_average)
# formula :  W = ((Rv + Cm)/(v+m))
    # W = weighted average
    # R = average vote for a movie 
    # v = number of votes for a movie
    # m = minimun votes required to be listed (currently 250)
    # C = mean vote accross the whole report
```

```python
#calculating the components
v = movies_cleaned_df['vote_count']
R = movies_cleaned_df['vote_average']
C = movies_cleaned_df['vote_average'].mean()
m = movies_cleaned_df['vote_count'].quantile(0.70)
    #m = total number of movies having vote of 70%
```

```python
#using the formula to calculate the weighted average
movies_cleaned_df['weighted_average'] = ((R*v)+(C*m))/(v+m)
movies_cleaned_df.head()
```

```python
#sort movies based on weighted average
movies_sorted_ranking = movies_cleaned_df.sort_values('weighted_average', ascending = False)
movies_sorted_ranking.head()
```

```python
#fetch top 20 movies with required info

req_info = ['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']

movies_sorted_ranking[req_info].head()
```

```python
# along with these two features(vote_count and vote_average), we can consider popularity also
# Recommendation based on scaled average weight and popularity

!pip3 install -U scikit-learn scipy matplotlib
from sklearn.preprocessing import MinMaxScaler
```

```python
#transform weighted_average and popularity columns (scaling down values (0 - 1))
scaling = MinMaxScaler()

movies_scaled_df = scaling.fit_transform(movies_cleaned_df[['weighted_average','popularity']])
movies_scaled_df # it is an numpy ndarray
```

```python
# create dataframe 
movies_normalized_df = pd.DataFrame(movies_scaled_df, columns=['weighted_average','popularity'])
movies_normalized_df.head()
```

```python
# combine this data with original data frame
movies_cleaned_df[['normalized_weighted_average','normalized_popularity']] = movies_normalized_df
movies_cleaned_df.head()
```

```python
#create a new column called score with 50% poportion

movies_cleaned_df['score'] = (movies_cleaned_df['normalized_weighted_average'] * 0.5) + (movies_cleaned_df['normalized_popularity'] * 0.5)

movies_cleaned_df.head()
```

```python
# sort based on score
movies_score_df = movies_cleaned_df.sort_values(['score'], ascending = False)
movies_score_df.head()
```
